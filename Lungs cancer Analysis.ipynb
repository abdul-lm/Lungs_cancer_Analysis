```python
data['GENDER']=data['GENDER'].replace({'M': 1, 'F': 2})
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
feature=['SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE','CHRONIC DISEASE','WHEEZING','ALCOHOL CONSUMING','COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']
x=data[feature]
y=data["LUNG_CANCER"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
log_pred=log.predict(x_test)
```

```python
log_pred
```

```python
log_acc=accuracy_score(y_test,log_pred)
print("Logistic Accuracy Score:","{:.2f}%".format(100*log_acc))
```

```python
from mlxtend.plotting import plot_confusion_matrix
```

```python
cm=confusion_matrix(y_test,log_pred)
plt.figure()
plot_confusion_matrix(cm,figsize=(12,8),hide_ticks=True,cmap=plt.cm.Blues)
plt.title("Logistic Regression - Confusion Matrix")
plt.xticks(range(2),["lungs Not Failed","lungs Fail"],fontsize=16)
plt.yticks(range(2),["lungs Not Failed","lungs Fail"],fontsize=16)
plt.show()
```
